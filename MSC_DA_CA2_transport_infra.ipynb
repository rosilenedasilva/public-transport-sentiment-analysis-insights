#### Dataset 4 : Gross domestic product (GDP)

https://data.oecd.org/gdp/gross-domestic-product-gdp.htm#indicator-chart

Gross domestic product (GDP) is the standard measure of the value added created through the production of goods and services in a country during a certain period. As such, it also measures the income earned from that production, or the total amount spent on final goods and services (less imports). While GDP is the single most important indicator to capture economic activity, it falls short of providing a suitable measure of people's material well-being for which alternative indicators may be more appropriate. This indicator is based on nominal GDP (also called GDP at current prices or GDP in value) and is available in different measures: US dollars and US dollars per capita (current PPPs). All OECD countries compile their data according to the 2008 System of National Accounts (SNA). This indicator is less suited for comparisons over time, as developments are not only caused by real growth, but also by changes in prices and PPPs.

```python
# Extract the year from the 'TIME' column and convert it to int64
car_regist['YEAR'] = car_regist['YEAR'].str.split('-').str[0].astype('int64')

# Renaming the DataFrame
newcar_regist = car_regist

# Display the new DataFrame, where the 'TIME' column only contains the year as an integer
newcar_regist.head()
```

The aggregation of passenger car registration data by averaging index values was chosen because it best reflects the annual trend of car registrations relative to the base year of 2015. This approach is consistent with traditional statistical practices and correlates with the indexed structure of the data by providing a clear, annual comparative indicator, which is critical for analysing year-over-year trends. 

```python
# Average the index 'Value' for each combination of 'LOCATION', 'INDICATOR', 'SUBJECT', 'MEASURE', and 'TIME'
newcar2_regist = newcar_regist.groupby(['COUNTRY', 'INDICATOR', 'SUBJECT', 'MEASURE', 'YEAR']).agg({'VALUE_PASSCAR': 'mean'}).reset_index()

# Checking the aggregated data
newcar2_regist.head()
```

Given that the datasets 'infra_invest' and 'passenger_transport' include the identical observations for the 'COUNTRY' and 'YEAR' columns, a more precise merge utilising these two keys was performed. This method will align the data more properly and avoid missing values. Because the datasets 1 and 2 have comparable columns and observations, they were joined using the procedure `on how=left`. The merging technique `on how=outer` was tested with the key columns COUNTRY', 'YEAR', and 'SUBJECT', however the outputs had too many missing data.

```python
# Merging the DataFrames on 'COUNTRY' and 'YEAR'
combined_df = pd.merge(newinfra_invest_df, newpassenger_transp, on=['COUNTRY', 'YEAR',], how='left')

# Display the first few rows of the combined DataFrame
combined_df.head(2)
```
